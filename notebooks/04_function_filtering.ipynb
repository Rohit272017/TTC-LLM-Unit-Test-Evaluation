```python
from pathlib import Path
import pandas as pd
import re
```

```python
functions_dir = Path(r"C:\Users\HOW1KOR\OneDrive - Bosch Group\Thesis\datasets\extracted\functions")

function_files = list(functions_dir.glob("*.cpp"))

print("Total functions:", len(function_files))
```

```python
keywords = [
    "if",
    "else",
    "switch",
    "case",
    ">=",
    "<=",
    "==",
    "return false",
    "return true",
    "throw",
    "invalid",
    "error"
]
```

```python
good_candidates = []

for file in function_files:

    text = file.read_text(errors="ignore")

    score = 0

    for keyword in keywords:
        score += text.count(keyword)

    if score >= 10 and len(text) < 15000:

        good_candidates.append({
            "filename": file.name,
            "score": score,
            "size": len(text)
        })

candidates_df = pd.DataFrame(good_candidates)

candidates_df = candidates_df.sort_values(
    by="score",
    ascending=False
)

candidates_df.head(20)
```

```python
candidate_name = candidates_df.iloc[0]["filename"]

sample = functions_dir / candidate_name

print(candidate_name)

print("\n===== FUNCTION =====\n")

print(sample.read_text()[:4000])
```
