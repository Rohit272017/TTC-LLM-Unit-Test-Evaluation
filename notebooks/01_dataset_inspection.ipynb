```python
from pathlib import Path

functions_dir = Path(
    r"C:\Users\HOW1KOR\OneDrive - Bosch Group\Thesis\datasets\extracted\functions"
)

human_tests_dir = Path(
    r"C:\Users\HOW1KOR\OneDrive - Bosch Group\Thesis\datasets\extracted\human_tests"
)

print(functions_dir.exists())
print(human_tests_dir.exists())
```

```python
for idx, row in df.iterrows():

    code = row["Code"]

    filename = f"function_{idx}.cpp"

    with open(functions_dir / filename, "w", encoding="utf-8") as f:
        f.write(code)

print("Functions extracted successfully.")
```

```python
for idx, row in df.iterrows():

    test_code = row["Unit Test - (Ground Truth)"]

    filename = f"human_test_{idx}.cpp"

    with open(human_tests_dir / filename, "w", encoding="utf-8") as f:
        f.write(test_code)

print("Human tests extracted successfully.")
```
